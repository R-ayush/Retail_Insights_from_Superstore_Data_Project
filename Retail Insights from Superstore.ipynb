```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
# Region-wise analysis
df.groupby("Region")["Sales"].sum().plot(kind='bar', ax=axes[0, 0], title='Region-wise Sales')
df.groupby("Region")["Profit"].sum().plot(kind='bar', ax=axes[0, 1], title='Region-wise Profit')

# Segment-wise analysis
df.groupby("Segment")["Sales"].sum().plot(kind='bar', ax=axes[0, 2], title='Segment-wise Sales')
df.groupby("Segment")["Profit"].sum().plot(kind='bar', ax=axes[1, 0], title='Segment-wise Profit')

# State-wise analysis
df.groupby("State")["Sales"].sum().plot(kind='bar', ax=axes[1, 1], title='State-wise Sales')
df.groupby("State")["Profit"].sum().plot(kind='bar', ax=axes[1, 2], title='State-wise Profit')

plt.tight_layout()
plt.show()
```
